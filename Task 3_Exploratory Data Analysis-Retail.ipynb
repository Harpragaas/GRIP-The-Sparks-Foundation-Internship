### 2. Exploring target variable

```python
# Creating the heatmap 
plt.figure(figsize=(5,5))
cor = store.corr()
sns.heatmap(cor,annot=True)
plt.show()
```

###### For the above problem statement : Profit is the target variable

##### From the correlation map it is observed that Sales, Quantity are positively correlated while Discount and Postal Code are negatively correlated

```python
#1 Profit
plt.figure(figsize = (8,8))
sns.distplot(store['Profit'])
plt.show()
```

##### The graph shows distribution of profit in the dataset, it is observed from the graph that the range of profit and loss is from 8000 to 6000 respectively from the sale of different items. It is also observed that majority of the products are sold at market retail price.

```python
#2 Sales
plt.figure(figsize = (8,8))
sns.distplot(store['Sales'],color = 'green')
plt.show()
```

##### The graph shows distribution of sales in the dataset, it is observed most sales have been made under 3000 however in some cases sales have been made close to approximately 24000.

```python
#3 Discount
plt.figure(figsize = (8,8))
store['Discount'].value_counts().plot.pie(autopct = '%1.1f%%')
plt.show()
```

##### The pie chart shows that approx 50% of the products are sold at market price,that is, without any discount. 37% of the products are sold at 20% discount followed by other discounts.

### 3. Quantative Analysis

Using Bivariate Anlaysis to find out relation between Profit (Target variable) and other variables

```python
# 1.Profit vs Sales 
plt.figure(figsize = [10,5])
plt.scatter(store.Profit , store.Sales , color = 'red')
plt.xlabel('Profit')
plt.ylabel('Sales')
plt.show()
```

##### Profit has been made where sales are bigger in number. Significant profit is made on sales more than 7000 while there are few instances of losses but they are significant. 

```python
# 2.Profit vs Discount
plt.figure(figsize = [10,5])
plt.scatter(store.Profit , store.Discount , color = 'green')
plt.xlabel('Profit')
plt.ylabel('Discount')
plt.show()
```

##### It is observed that when there is no or less discount, the margin of profit is more as compared to discount of 50% led to losses 

```python
# 3.Profit vs Quantitiy
plt.figure(figsize = [10,5])
plt.scatter(store.Profit , store.Quantity , color = 'blue')
plt.xlabel('Profit')
plt.ylabel('Quantity')
plt.show()
```

##### It can be infered that Profits are high when products are bought in bulk 

```python
# 4.Profit vs Segment 
plt.figure(figsize = [10,5])
sns.violinplot(x="Profit", y="Segment", data=store)
plt.show()
```

#####  It is seen that data distribution is most in consumer segment followed by corporate and home office. Proportion of profit and loss is almost identical in Consumer segment while corporate segment makes the most profit than the other two segments. 

```python
# 5.Profit vs Region 
plt.figure(figsize = [10,5])
sns.violinplot(x="Profit", y="Region", data=store)
plt.show()
```

##### From the graph it can be observed that East region made the most losses for the store while sales in the Central region were the most profitable. West followed the central region in terms of profit while the south region showed equla proportion of profit and losses.

```python
# 6.Profit vs Category 
plt.figure(figsize = [10,5])
sns.violinplot(x="Profit", y="Category", data=store)
plt.show()
```

##### Data distribution is highest in furtinure followed by office supplies and technology. Profit and Loss is the highest in technology category but in equal proportion

```python
# 7.Profit vs Ship Mode
plt.figure(figsize = [10,5])
sns.violinplot(x="Profit", y="Ship Mode", data=store)
plt.show()
```

##### It is observed that Standard class has highest proportion of profit and loss. First class makes the most profit however it is less used as the shipping mode. Second Class was the most popular shipping method (after the standard class) makes more profit while same day shipping lead to losses for the store.   

#### Problem Statement 1
#### Insights : Weak areas to work on

1. There is no data available of the North Region, we should also include this data for better analysis.

2. The store should focus on the Eastern part of the country and try to perform a survey in order to understand the customer demand in that region.
